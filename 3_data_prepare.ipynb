```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # showing and rendering figures
# io related
# from skimage.io import imread
# import os
# from glob import glob
# # not needed in Kaggle, but required in Jupyter
%matplotlib inline 
```

```python
data_df = pd.read_csv('dataset.csv')
data_df['x'].hist(figsize = (10, 5))
```

```python
data_df['x_category'] = pd.cut(data_df['x'], 3)
new_data_df = data_df.groupby(['x_category']).apply(lambda x: x.sample(500, replace = True)).reset_index(drop = True)
new_data_df['x'].hist(figsize = (10, 5))
```

```python
new_data_df.to_csv("dataset2.csv")
```
